#EXPLORATORY DATA ANALYSIS: 

```python
mu = 0
sigma = 1
x_norm = np.arange(-5, 5, 0.1)
y_norm = stats.norm.pdf(x_norm, 0, 1)

plt.plot(x_norm, y_norm)
plt.title('Normal: $\mu$=%.1f, $\sigma^2$=%.1f' % (mu, sigma))
plt.xlabel('x')
plt.ylabel('Probability density')
plt.show()

```

Function Description:
This function utilizes plot, mu, sigma and the scipy.stats library to access the 'pdf' density function to display the normal distribution of the Probability Density Function. 

Exploratory Analysis:
The normal distribution probability density, the plot appears to have a Gaussian distribution. Probability Density increases after -4.0, reaches a peak at 0 and decreases afterwards.It is not known if the distribution continues to infinity. The probability density can work with cases regarding height among teenagers. X and Y-values would have to be changed to realistically accomodate the height differences.

```python
mu = 0
sigma = 1
x_norm = np.arange(-4, 4, 0.1)
y_norm = stats.norm.pdf(x_norm, 0, 1)
df1 = stats.t.pdf(x_norm,4)
df2 = stats.t.pdf(x_norm, 8)
df3 = stats.t.pdf(x_norm, 12)
df4 = stats.t.pdf(x_norm, 30)

plt.plot(x_norm, y_norm)
plt.plot(x_norm,df1)
plt.plot(x_norm,df2)
plt.plot(x_norm,df3)
plt.plot(x_norm, df4)
plt.title('Density T-Distribution: $\mu$=%.1f, $\sigma^2$=%.1f' % (mu, sigma))
plt.xlabel('x')
plt.ylabel('Probability density')
plt.plot(x_norm, y_norm, label='normal')
plt.plot(x_norm, df1, label='df1')
plt.plot(x_norm, df2, label='df2')
plt.plot(x_norm, df3, label='df3')
plt.plot(x_norm, df4, label='df4')
plt.legend();
plt.show()
```

Function Description:
This function utilizes plot to display the various degrees of freedom of pdf, or the Probability Density Function. Each degree of freedom is given a specific color and is set by its own variable df1,df2,df3,df4, which are plotted independently.

Exploratory Analysis:
Regarding the various degrees of freedom and their differences from the normal distribution, the various degrees of freedom density is larger before and after reaching the mean of 0 compared to the normal distribution. Variance seems to be low considering that all probabilities start clustering at -2 (x-axis) and peak at 0. 

```python
mu = 0
sigma = 1
x_norm = np.arange(-5, 5, 0.1)
y_norm = stats.norm.cdf(x_norm, 0, 1)

plt.plot(x_norm, y_norm)
plt.title('Normal: $\mu$=%.1f, $\sigma^2$=%.1f' % (mu, sigma))
plt.xlabel('x')
plt.ylabel('Cumulative Distribution')
plt.show()
```

Function Description:
This function utilizes plot, mu, sigma and the scipy.stats library to access the 'cdf' density function to display the normal Cumulative Distribution Function. 

Exploratory Analysis:
Based on the normal distribution for the Cumulative Distribution Function, probability(Y-axis) increases from 0 to 1 from left to right. If the plot is continous and infinite, the right side of the plot will likely contain a random variable (x) that will be less than or equal to X. Although, It is not known if both head and tail probabilities will continue infinitely. It appears that the probability will remain constant at 0, if the random variable x decreases past -2, or remain constant at 1 if x increases past 4.

```python
mu = 0
sigma = 1
x_norm = np.arange(-3, 3, 0.1)
y_norm = stats.norm.cdf(x_norm, 0, 1)
df1 = stats.t.cdf(x_norm,4)
df2 = stats.t.cdf(x_norm, 8)
df3 = stats.t.cdf(x_norm, 12)
df4 = stats.t.cdf(x_norm, 30)

plt.plot(x_norm, y_norm)
plt.plot(x_norm,df1)
plt.plot(x_norm,df2)
plt.plot(x_norm,df3)
plt.plot(x_norm, df4)
plt.title('CDF T-Distribution: $\mu$=%.1f, $\sigma^2$=%.1f' % (mu, sigma))
plt.xlabel('x')
plt.ylabel('Cumulative Distribution')
plt.plot(x_norm, y_norm, label='normal')
plt.plot(x_norm, df1, label='df1')
plt.plot(x_norm, df2, label='df2')
plt.plot(x_norm, df3, label='df3')
plt.plot(x_norm, df4, label='df4')
plt.legend();
plt.show()
```

Function Description:
This function utilizes plot to display the various degrees of freedom of cdf, or the Cumulative Probability Function. Each degree of freedom is given a specific color and is set by its own variable df1,df2,df3,df4, which are initialized and plotted independently.

Exploratory Analysis:
The T distribution of the function displays larger probabilities for all degrees of freedom compared to the normal distribution before 0. T distribution probabilities seem to converge at 0 on the x-axis, then probability increases to become less than or equal to 1 on the Y-axis.

```python
x = np.linspace(-5, 5, 5000)
mu = 0
sigma = 1

y_ppf = stats.norm.ppf(x, mu, sigma)
plt.xlabel('x')
plt.ylabel('Quantile')
plt.plot(x, y_ppf, label='ppf')
```

Function Description:
This function utilizes plot, mu, sigma and the scipy.stats library to access the 'ppf' Quantile function to display the normal Inverse Cumulative Distribution Function. 

Exploratory Analysis:
Based on the plot of the normal inverted CDF, unlike the probabilities from the normal non-inverted CDF, the probability appears to increase at the random x variable 1 and decrease at -1. It is also not known if the line of probability on the plot is infinity. This Quantile function displays a low variance with probabilities near the mean of 0.5.

```python
x_norm = np.linspace(-5, 5, 5000)
mu = 0
sigma = 1

y_norm = stats.norm.ppf(x, mu, sigma)
##plt.plot(x, y_ppf, label='ppf')

##y_norm = stats.norm.ppf(x_norm, 0, 1)
df1 = stats.t.ppf(x_norm,4)
df2 = stats.t.ppf(x_norm, 8)
df3 = stats.t.ppf(x_norm, 12)
df4 = stats.t.ppf(x_norm, 30)

plt.plot(x_norm, y_norm)
plt.plot(x_norm,df1)
plt.plot(x_norm,df2)
plt.plot(x_norm,df3)
plt.plot(x_norm, df4)
plt.title('CDF T-Distribution: $\mu$=%.1f, $\sigma^2$=%.1f' % (mu, sigma))
plt.xlabel('x')
plt.ylabel('Cumulative Distribution')
plt.plot(x_norm, y_norm, label='normal')
plt.plot(x_norm, df1, label='df1')
plt.plot(x_norm, df2, label='df2')
plt.plot(x_norm, df3, label='df3')
plt.plot(x_norm, df4, label='df4')
plt.legend();
plt.show()
```

Function Description:
This function utilizes plot to display the various degrees of freedom of ppf, or the Quantile Function. Each degree of freedom is given a specific color and is set by its own variable df1,df2,df3,df4, which are initialized and plotted independently.

Exploratory Analysis:
The T distribution of the various freedoms seems to follow the same path of probability as the normal distribution displayed on the plot. The random x variables of the various freedoms seems to be greater than 0 on the x-axis before converging at the mean and becomes smaller afterwards at 1. The 4th degree of freedom has lower variance compared to the other degrees.
